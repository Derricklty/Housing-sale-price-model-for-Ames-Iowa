# Problem Statement

Ames is a city in Story County, Iowa, United States, located approximately 48 km north of Des Moines in central Iowa. The city is best known as the home of Iowa State University (ISU), with leading agriculture, design, engineering, and veterinary medicine colleges. As of 2019, the city had a population of 66,258 (including 33,391 students from ISU which makes up close to half of the city's population. Given its prime location, Ames was ranked ninth on CNNMoney's "Best Places to Live" list in 2010 ([*Source: Wikipedia*](https://en.wikipedia.org/wiki/Ames,_Iowa))


As a junior data scientist in Ames Estates, one of the biggest real estate companies in Ames, you have been tasked by your CEO to develop an appropriate regression model to predict sale prices for properties in the city. With this model, it is your boss' wish that your company could make well-informed pricing decisions (for the properties our company is selling) without having to rely too heavily on the opinions and expertise of your realtors. The model will also serve as a systematic, data-driven way of estimating a sale price which we can share with house owners when they entrust us with the responsibility of selling their properties.

### Assigning numerical values to ordinal variables

```python
'''
For ordinal variables, we have to assign numeric values according to the data dictionary 
for our machine to be able to read them. To do so, we create a function for those with 
similar mapping indexes.
#Reference: https://towardsdatascience.com/ways-to-handle-categorical-data-before-train-ml-models-with-implementation-ffc213dc84ec
'''
ordinal_vars = ["exter_qual", "exter_cond", "bsmt_qual", "bsmt_cond", "heating_qc", "kitchen_qual","garage_qual", "garage_cond", "fireplace_qu","pool_qc"]

for col in ordinal_vars:
    train_data[col] = train_data[col].map({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
    
for col in ordinal_vars:
    test_data[col] = test_data[col].map({'NA':0,'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5})
    
```

```python
# As Lot Shape is an ordinal variable, we will assign numeric values according to the data dictionary
train_data["lot_shape"] = train_data["lot_shape"].map({'Reg':0,'IR1':1,'IR2':2,'IR3':3})
test_data["lot_shape"] = test_data["lot_shape"].map({'Reg':0,'IR1':1,'IR2':2,'IR3':3})
```

- As a guide, we will use the correlation heatmap to identify variables that are highly correlated with sale price

```python
#Correlation heatmap of all variables against target

plt.figure(figsize=(10,20))
sns.heatmap(train_data.corr()[['saleprice']].sort_values("saleprice", ascending = False), vmin= -1, vmax=1, annot=True)
plt.title('Correlations of Variables with Sale Price', fontdict={'fontsize':14})
plt.ylabel('Features');
```

### Observations
From the correlation heatmap, we observed that "overall_qual", "exter_qual", "kitchen_qual", "gr_liv_area", "garage_area", "garage_cars", "total_bsmt_sf", "1st_flr_sf", "bsmt-qual", "year_built","garage_finish", "fireplace_qu", "year_remod/add", "full_bath", "mas_vnr_area", "totrms_abvgrd" are highly correlated with Sale Price. These variables have correlation > 0.5 with sale price

```python
'''
For the variables that are highly correlated to sale price, we might not want to include 
all of them into our regression as there may be multicollinearity. 
To check for multicollinearity, we will draw a simple heatmap for these variables
'''

#Creating a subset of the dataset based on the above heatmap
train_data_eda = train_data[["overall_qual", "exter_qual", "kitchen_qual", "gr_liv_area", 
                             "garage_area", "garage_cars", "total_bsmt_sf", 
                             "1st_flr_sf", "bsmt_qual", "year_built", "garage_finish",
                             "year_remod/add",  "full_bath",
                             "mas_vnr_area", "totrms_abvgrd"]]
                             
                             
plt.figure(figsize=(12, 12))
sns.heatmap(train_data_eda.corr(),annot= True, mask = np.triu(train_data_eda.corr()))
plt.title('Correlations of Variables with Sale Price', fontdict={'fontsize':14})
plt.ylabel('Features');
```

### Observations

To avoid the issue of multicollinearity, we will pick "overall_qual", "gr_liv_area", "garage_area", "total_bsmt_sf", "bsmt_qual", "year_built","garage_finish", "full_bath", "mas_vnr_area" while disregarding the rest of the variables. ## Modeling 

### Model 1
- We start off with a simple linear regression with all numerical and dummy variables as the independent variables. This serves as a sanity check that we should never include all available variables into our model.

```python
'''
As we have added the log of some continuous variables to train_data, we need to create another dataset without
the logarithmic forms to prevent multicollinearity
'''

train_data_model_1 = train_data.drop(columns = ["log_saleprice", "log_gr_liv_area", "log_garage_area",
                                              "log_total_bsmt_sf", "log_mas_vnr_area"])
```

```python
#We create a new dataset which include all the numerical and dummy variables
model_1 = pd.get_dummies(train_data_model_1, columns = nominal_vars_train, drop_first= True)

#Saving a copy of the data
model_1.to_csv("../datasets/model_1.csv", index=False)
```

```python
# Defining our X and y
X_model_1 = model_1.drop(columns = ['id', 'pid', 'saleprice'])
y_model_1 = model_1["saleprice"]
```

```python
#Doing train-test split with random seed of 42
X_train, X_test, y_train, y_test = train_test_split(X_model_1, y_model_1, test_size = 0.3, random_state=42)
```

```python
# To remove the effects of large values on our model, we standard scale all our variables
ss = StandardScaler()

Z_train = ss.fit_transform(X_train)
Z_test = ss.transform(X_test)
```

```python
#Instantiate and fit our model
lr = LinearRegression()
lr.fit(Z_train, y_train)
```

```python
# Train, test and cross val scores

train_score = lr.score(Z_train, y_train)
print(f"Train score is {train_score}")


test_score = lr.score(Z_test, y_test)
print(f"Test score is {test_score}")

crossval_score = cross_val_score(lr, Z_train, y_train, cv = 5).mean()
print(f"Cross val score is {crossval_score}")

predictions = lr.predict(Z_test)
```

```python
# Check the RMSE on the training and testing sets.

print(f'RMSE on training set: {mean_squared_error(y_train, lr.predict(Z_train), squared = False)}')
print(f'RMSE on testing set: {mean_squared_error(y_test, predictions, squared = False)}')
```

### Observation for model 1

From the various test scores, they seem suggest that there is overfitting (training scores and training RMSE are performing way better than their counterparts for the test set). This also illustrates why it is important that we curate our list of independent variables (which we did via the correlation heatmap earlier).

### Model 4

LASSO regression using same variables in Model 2

```python
# LASSO model
#Doing train-test split with random seed of 42
X_train, X_test, y_train, y_test = train_test_split(X_model_2, y_model_2, test_size = 0.3, random_state=42)
```

```python
#Instatiate the model
lasso_model = Lasso(alpha=1)

# Fit
lasso_model.fit(Z_train, y_train)

# Evaluate model using R2.
print(f"LASSO model train score is {lasso_model.score(Z_train, y_train)}")
print(f"LASSO model test score is {lasso_model.score(Z_test, y_test)}")

crossval_score = cross_val_score(lasso_model, Z_train, y_train, cv = 5).mean()
print(f"Cross val score is {crossval_score}")
```

```python
# Here is the optimal value of alpha
lasso_cv.alpha_
```

In terms of the interpretation of coefficients, they are as such: 
- for continuous/discrete/ordinal variables, a one unit increase in [variable] will result in a beta increase/decrease in the sale price (holding everything else constant). For e.g. increasing above grade living area (grv_liv_area) by one square feet increases sale price by ~$27,000.
- for dummy variables, beta represents the addition increase/decrease in sale price over and above the base dummy (category that is dropped as part of drop_first= True; holding everything else constant). For example, a property in Northridge Heights (neighborhood_nridgHt) will see its sale price to be ~$$8,500 higher than that of a property in Bloomington Heights (base category for our neighborhood dummies).
